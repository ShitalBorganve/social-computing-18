# NetworkX ... Python Library for Network Analysis

## Part 1: Understanding graph basics using NetworkX library

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
G.add_node(1)
```

```python
G.add_nodes_from([2,3,4])
```

```python
G.nodes()
```

```python
G.add_edge(1,2)
```

```python
G.add_edges_from([(1,3),(2,4),(3,4),(1,4)])
```

```python
G.edges()
```

```python
G.number_of_nodes()
```

```python
G.number_of_edges()
```

```python
import matplotlib.pyplot as plt
```

```python
nx.draw_networkx(G)
```

```python
G.add_node(5)
```

```python
G.add_edge(4,5)
```

```python
nx.draw_networkx(G)
```

```python
nx.draw_networkx(G,nodecolor='r',edge_color='b')
```

```python
DG = nx.DiGraph()
```

```python
DG.add_nodes_from([1,2,3,4,5,6,7])
```

```python
DG.add_weighted_edges_from([(1,2,0.5),(2,3,0.63),(1,4,0.32),(2,4,0.15),(6,5,0.45),(7,6,0.89)])
```

```python
nx.draw_networkx(DG,nodecolor='r',edge_color='b')
```
